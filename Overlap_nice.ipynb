```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
m1 = 1.0
m2 = 1.5
std1 = 0.3
std2 = 0.1

x = np.linspace(0.2,4,10000)
plot1=plt.plot(x,norm.pdf(x,m1,std1))
plot2=plt.plot(x,norm.pdf(x,m2,std2))

xd = np.vstack((norm.pdf(x,m1,std1),norm.pdf(x,m2,std2)))
print(xd.shape)
np.min(xd, axis=0)

plt.plot(x,np.min(xd, axis=0))

plt.fill(x,np.min(xd, axis=0), facecolor='green', alpha=0.5)
```
